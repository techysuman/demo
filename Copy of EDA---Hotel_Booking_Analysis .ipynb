18. **previous_cancellations** : *Number of previous bookings that were cancelled by the customer prior to the current booking*

19. **previous_bookings_not_canceled** : *Number of previous bookings not cancelled by the customer prior to the current booking*

22. **booking_changes** : *Number of changes/amendments made to the booking from the moment the booking was entered on the PMS until the moment of check-in or cancellation*

23. **deposit_type** : *Indication on if the customer made a deposit to guarantee the booking.*

24. **agent** : *ID of the travel agency that made the booking*

25. **company** : *ID of the company/entity that made the booking or responsible for paying the booking.*

26. **days_in_waiting_list** : *Number of days the booking was in the waiting list before it was confirmed to the customer*

27. **customer_type** : *Type of booking, assuming one of four categories*


28. **adr** : *Average Daily Rate as defined by dividing the sum of all lodging transactions by the total number of staying nights*

29. **required_car_parking_spaces** : *Number of car parking spaces required by the customer*

30. **total_of_special_requests** :* Number of special requests made by the customer (e.g. twin bed or high floor)*

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
## Copy the dataset
df = hotel_booking_df.copy()
````

````python
df.info()
````

````python
# Write your code to make your dataset analysis ready.
df= df.drop_duplicates()

# data set reduced
df.shape
````

````python
#replace Undefined, BB, FB, HB, SC to its meaning.
df['meal'].replace(['Undefined', 'BB', 'FB', 'HB', 'SC'],
                   [ 'No Meal', 'Breakfast', 'Full Board', 'Half Board', 'Self Catering'],
                   inplace = True)
````

````python
# We can drop 4 rows containing null values from chidren column
# It will not affect our result too much
df = df.dropna(subset = ['children'])
````

````python
# For country we will fill missing values with the mode frequent value.
df['country'].fillna(df['country'].mode()[0], inplace = True)
````

````python
## If no id of agent or company is null, just replace it with 0
df[['agent','company']] = df[['agent','company']].fillna(0.0)
````

````python
## convert datatype of these columns from float to integer
df[['children', 'company', 'agent']] = df[['children', 'company', 'agent']].astype('int64')
````

````python
#There are many rows that have zero guests including adults, children and babies. These type of rows does not make sense
df[(df.adults+df.babies+df.children)==0].shape
````

````python
## Drop Rows where there is no adult, baby and child because it simply means no bookings were made
df = df.drop(df[(df.adults+df.babies+df.children)==0].index)
````

````python
# Adding total staying days in hotels
df['total_stay'] = df['stays_in_weekend_nights']+df['stays_in_week_nights']

# Adding total people number as column, i.e. total types of person = num of adults + children + babies
df['total_people'] = df['adults']+df['babies']+df['children']
````

````python
#turn column into datetime data type
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'], format = '%Y-%m-%d')
````

````python
df.info()
````

### What all manipulations have you done and insights you found?

The manipulations we have done and the insights were found, are as follows:-


*   I found that there were four columns containing null values. So we had Null values in columns- Company, Agent, Country and Children.

  1.   For company and agent, i have filled the missing values with 0

  2.   For country column, i have fill missing values with object 'Mode of the country column'(Assuming while collecting data country was not found so changing it with the mode will make a lot of sense.)

  3.   As the count of missing values in Children Column was only 4, so I dropped those rows.


*   This dataset was also containing duplicate values, so duplicate values was dropped.

*   I found that there were some rows in which the combining values of adults, babies and childrens was 0, so this simply means there were no guests as 0 indicates presence of none. So, there were no bookings made. As a result, i simply dropped the rows where combining values of adults, babies and children columns was 0.

*   The data type of 'reservation_status_date' column was object type, so it was changed to date type format for better use.

*   There were two new columns that was added, one is 'total_people' and other is 'total_stay'.

#### Chart - 5 ---> Month wise booking trend analysis.

````python
# Using groupby on arrival_date_month and taking the hotel count
bookings_by_months_df = hotel_booking_df.groupby(['arrival_date_month'])['hotel'].count().reset_index().rename(columns = {'hotel':'Counts'})

# Creating list of months in order
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

# Creating dataframe which will map the order of above months list without changing its values
bookings_by_months_df['arrival_date_month'] = pd.Categorical(bookings_by_months_df['arrival_date_month'], categories = months, ordered = True)

# Sorting by arrival_date_month
bookings_by_months_df = bookings_by_months_df.sort_values('arrival_date_month')

bookings_by_months_df
````
